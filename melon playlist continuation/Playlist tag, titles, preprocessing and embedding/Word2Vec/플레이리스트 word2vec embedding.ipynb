```python
# 데이터셋 만들기
def get_dic(data):
    song_dic = {}
    title_tags_dic = {}
    for q in tqdm(data):
        song_dic[str(q['id'])] = q['songs']
        title_tags_dic[str(q['id'])] = q['title_tags']
    total = list(map(lambda x: list(x['songs']) + list(x['title_tags']), data))
    total = [x for x in total if len(x)>1]
    return song_dic, title_tags_dic, total
```

```python
song_dic, title_tags_dic, total = get_dic(train_df)
```

```python
from gensim.models.word2vec import Word2Vec
from gensim.models.callbacks import CallbackAny2Vec
import logging
```

```python
model_song = Word2Vec(
                 # epochs = 5, # neural network 학습할 때 전체 training data를 몇번 돌리는지
                 vector_size = 100, # 임베딩을 하는 벡터의 사이즈 (hidden layer의 size)
                 min_count = 1, # 단어가 최소 등장해야될 횟수
                 # 단어가 겹치는게 많아야 학습이 잘됨, 너무 적으면 의미가 없음, data_preprocessing
                 workers = 4,
                 sg = 1, # skipgram을 사용한다면 1, 0을하면 CBOW가 됨
                 hs = 0, # negative sampling을 사용한다면 0 (hierarchical softmax)
                 negative = 50, # 1개의 positive sample당 negative sample의 개수
                 window = 300 # window를 크게 잡는 이유는 문장 내 모든 아이템끼리 학습이 되어야 하기 때문임
                 # 여기서 word2vec의 input은 집합이 아니라 list임, 그래서 window size를 크게 해야 모든 조합들이 
                 # 다 학습 데이터에 들어감
                 # item2vec의 집합으로 만들기 위해서 window size를 크게 함
                )

print(model_song)
# word_vectors = model.wv
```

```python
logging.basicConfig(format="%(asctime)s : %(levelname)s : %(message)s", level=logging.INFO)

class Callback(CallbackAny2Vec):
    def __init__(self):
        self.epoch = 1
        self.training_loss = []

    def on_epoch_end(self, model):
        loss = model.get_latest_training_loss()
        if self.epoch == 1:
            current_loss = loss
        else:
            current_loss = loss - self.loss_previous_step
        print(f"Loss after epoch {self.epoch}: {current_loss}")
        self.training_loss.append(current_loss)
        self.epoch += 1
        self.loss_previous_step = loss
```

```python
logging.disable(logging.NOTSET) # enable logging
t = time()

model_song.build_vocab(total)
```

```python
model_song.corpus_count
```

```python
%%time
logging.disable(logging.INFO) # disable logging
callback = Callback() # instead, print out loss for each epoch
t = time()

model_song.train(total,
            total_examples = model_song.corpus_count,
            epochs = 10,
            compute_loss = True,
            callbacks = [callback]) 

model_song.save("title_tags2vec(word2vec).model")
print(model_song)
```
